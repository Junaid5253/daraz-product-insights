```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("clean_products.csv")
df.head()
```

```python
df["Revenue"] = df["Price"] * df["Units Sold"]
```

```python
print("Total Products:", len(df))
print("Total Revenue:", df["Revenue"].sum())
print("Average Price:", df["Price"].mean())
print("Average Units Sold:", df["Units Sold"].mean())
print("Categories:", df["Category"].unique())
```

```python
top_products = df.sort_values("Units Sold", ascending=False).head(10)
top_products[["ProductName", "Category", "Units Sold", "Price", "Revenue", "Rating"]]
```

```python
top_brands = df.groupby("Brand")["Revenue"].sum().sort_values(ascending=False).head(10)
top_brands.plot(kind="bar", figsize=(10,5), color="skyblue", title="Top 10 Brands by Revenue")
plt.ylabel("Revenue (PKR)")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df["Price"], bins=50, kde=False)
plt.title("Price Distribution")
plt.xlabel("Price (PKR)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Price", y="Units Sold", hue="Category", data=df)
plt.title("Units Sold vs Price by Category")
plt.xlabel("Price (PKR)")
plt.ylabel("Units Sold")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[["Price", "Units Sold", "Revenue"]].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

# 11. Key Insights
- Products with moderate prices often sell the most.
- High-rated products usually generate higher revenue.
- Certain brands dominate in total revenue.
- Units Sold and Price show weak negative correlation, highlighting the price sensitivity.
